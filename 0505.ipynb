```python
import matplotlib.pyplot as plt
import numpy as np#矩陣函式庫
plt.plot([1,2,3,4],[0,0.3,0.6,0.9],'gx')#綠x標記
plt.plot([1,2,3,4],[0,0.3,0.6,0.9],'r--')#紅虛線
#取亂數
X = 1 + np.arange(30) / 10
delta=np.random.uniform(low = -0.1,high = 0.1,size = (30,))
#代公式
Y = 0.3*X - 0.3 + delta
plt.plot(X,Y,'bo')#藍點
plt.ylabel('X')#Y文字
plt.xlabel('Y')#X文字
plt.show()
```

```python
import pandas as pd
from sklearn import linear_model#繪圖函式庫
import matplotlib.pyplot as plt
#read data
dataframe = pd.read_fwf('brain_body.txt')#txt的數據
x_values = dataframe[['Body']] #身體的數據
y_values = dataframe[['Brain']] #大腦的的數據
#訓練

body_reg = linear_model.LinearRegression()#線性迴歸模型

body_reg.fit(x_values,y_values)
pre = body_reg.predict(x_values)

#訓練預測
print(body_reg.predict(pd.DataFrame(data=[[170]])))
#圖形化
plt.scatter(x_values,y_values)
plt.plot(x_values,pre)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets,linear_model
#取得數據

diabetes = datasets.load_diabetes()

#只取第三個特徵值BMI
diabetes_X = diabetes.data[:,np.newaxis,2]

#切分特徵值BMI
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]

#切分答案
diabetes_y_train = diabetes.target[:-20]
diabetes_y_test = diabetes.target[-20:]
```

```python
regr = linear_model.LinearRegression()
regr.fit(diabetes_X_train,diabetes_y_train)

print('Coefficients: \n',regr.coef_)

print("Mean squared error: %.2f"
     % np.mean((regr.predict(diabetes_X_test) - diabetes_y_test) ** 2))

print('Variance score: %.2f' % regr.score(diabetes_X_test,diabetes_y_test))

plt.scatter(diabetes_X_test,diabetes_y_test,color='black')
plt.plot(diabetes_X_test,regr.predict(diabetes_X_test),color='blue',
         linewidth=3)

plt.xticks(())
plt.yticks(())
plt.show()
```
